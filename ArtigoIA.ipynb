#Importando Bibliotecas

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
```

```python

# Construção do modelo machine learning
# Treino 75% dados - fit
# Teste 25% - predict -> Acuracia(verificao se a bom ou ruim o modelo que foi treinado)
# Se ACURACIA boa JOGA O MODELO NO MUNDO!!!
url = '/content/plant_growth_data.csv'
df = pd.read_csv(url)
df.head(10)
```

```python
# prompt: Usando o DataFrame df: Passar os dados de ingles para português

# Renomeando colunas
df = df.rename(columns={
    'Soil_Type': 'Tipo_Solo',
    'Sunlight_Hours': 'Horas_Sol',
    'Water_Frequency': 'Frequencia_Agua',
    'Fertilizer_Type': 'Tipo_Fertilizante',
    'Temperature': 'Temperatura',
    'Humidity': 'Umidade',
    'Growth_Milestone': 'Marco_Crescimento'
})

# Traduzindo valores categóricos
df['Tipo_Solo'] = df['Tipo_Solo'].replace({
    'loam': 'barro',
    'sandy': 'arenoso',
    'clay': 'argila'
})

df['Frequencia_Agua'] = df['Frequencia_Agua'].replace({
    'bi-weekly': 'quinzenal',
    'weekly': 'semanal',
    'daily': 'diariamente'
})

df['Tipo_Fertilizante'] = df['Tipo_Fertilizante'].replace({
    'chemical': 'quimico',
    'organic': 'organico',
    'none': 'nenhum'
})

df.head(10)
```

## Transformação dos dados Qualitativos em quantitativos

```python
# prompt: Usando o DataFrame df: Tranforme as colunas qualitativas em quantitativas onde, barro = 0 , arenoso = 1 e argila =2

# Mapeando os valores qualitativos para quantitativos na coluna 'Tipo_Solo'
mapeamento_solo = {'barro': 0, 'arenoso': 1, 'argila': 2}
df['Tipo_Solo'] = df['Tipo_Solo'].map(mapeamento_solo)
mapeamento_frequencia_agua = {'quinzenal': 0, 'semanal': 1, 'diariamente': 2}
df['Frequencia_Agua'] = df['Frequencia_Agua'].map(mapeamento_frequencia_agua)
mapeamento_fertilizante = {'quimico': 0, 'organico': 1, 'nenhum': 2}
df['Tipo_Fertilizante'] = df['Tipo_Fertilizante'].map(mapeamento_fertilizante)
df.head(10)
```

```python
x = df[['Tipo_Solo','Horas_Sol','Frequencia_Agua','Tipo_Fertilizante','Temperatura','Umidade']] # Use double brackets to select multiple columns
y = df['Marco_Crescimento']
```

```python
seed = 35
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.25, random_state=seed,stratify=y)

print('Treinamento com %d elementos e testarmos com %d elementos' % (len(train_x), len(test_x)))
```

```python
model = LinearSVC() # Cria o modelo
model.fit(train_x, train_y) # treino

result = model.predict(test_x) # teste

acuracia = accuracy_score(test_y, result) # compara vetores
acuracia_model = model.score(test_x, test_y) # roda teste

print("Acuracia_model: ", (acuracia_model*100), "%")
```

```python
from sklearn.dummy import DummyClassifier

dummy_stratified = DummyClassifier()
dummy_stratified.fit(train_x, train_y)
acuracia_dummy = dummy_stratified.score(test_x, test_y)

print("Acuracia Dummy: ", (acuracia_dummy*100), "%")
```

```python
# prompt: Melhore a acuracia do modelo

from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler

# Normalizando os dados
scaler = StandardScaler()
train_x_scaled = scaler.fit_transform(train_x)
test_x_scaled = scaler.transform(test_x)

# Testando diferentes kernels
for kernel in ['linear', 'poly', 'rbf', 'sigmoid']:
  model = SVC(kernel=kernel)
  model.fit(train_x_scaled, train_y)
  acuracia_model = model.score(test_x_scaled, test_y)
  print(f"Acurácia com kernel {kernel}: {(acuracia_model*100):.2f}%")

# Ajustando hiperparâmetros com GridSearchCV (pode demorar um pouco)
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [0.1, 1, 10], 'gamma': [0.1, 1, 10]}
grid_search = GridSearchCV(SVC(), param_grid, cv=5)
grid_search.fit(train_x_scaled, train_y)

# Avaliando o melhor modelo
best_model = grid_search.best_estimator_
acuracia_best_model = best_model.score(test_x_scaled, test_y)
print(f"Acurácia do melhor modelo: {(acuracia_best_model*100):.2f}%")
print(f"Melhores parâmetros: {grid_search.best_params_}")
```
